```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
std = pd.read_excel('std_marks_sheet.xlsx')
```

```python
std.head()
```

```python
A = std[std.Grade == "A"]
```

```python
B = std[std.Grade == "B"]
```

```python
std=pd.concat({"Grade_A": A, "Grade_B": B})
```

```python
std
```

```python
std.isna().sum()
```

```python
sns.barplot(x="Grade",y="Total",data=std)
plt.title("STD GRADE")
plt.show()
```

```python
plt.figure()
sns.barplot(x="ML", y="pandas", data=std)
plt.title("Pandas vs numpy")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.hist(std["roll num"], bins=54)
plt.title("Grade")
plt.xlabel("DL")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="numpy", y="Grade", data=std)
plt.title("NAME VS GRADE")
plt.show()
```

```python
std.to_csv('STUDENT.csv',index=False)
```
